```python
#в файле "Подарочный ролл по акции за 990р" долны отсутствовать пиццы и сеты
import pandas as pd
import numpy as np 
df_990 = pd.read_excel('Подарочный ролл по акции за 990р.xlsx')
df_990 = df_990[~pd.isna(df_990['Блюдо'])]
df_990['Номер чека']  = df_990['Номер чека'].ffill()

temp= df_990['Блюдо'].unique()

list_exclusion = [
'Кока-кола 0.9л', 
'Сок Santal Яблочный 1л',  
'Кока-кола 2л',  
'Сок Santal Вишня 1л',  
'Сок Santal Апельсиновый 1л']

list_gifts = [
'2 Ролла в подарок по Акции', 
' Ролл в подарок', 
'3 Ролла в подарок по Акции']

df_990 = df_990[~df_990['Блюдо'].isin(list_exclusion)]

list = []

for i in df_990['Номер чека'].unique():
    gifts_count= 0 
    sum_check = df_990[df_990['Номер чека'] == i]["Сумма со скидкой, р."].sum()
    gifts_count = gifts_count + df_990[df_990['Номер чека'] == i]["Блюдо"].values.tolist().count(' Ролл в подарок')
    gifts_count = gifts_count + df_990[df_990['Номер чека'] == i]["Блюдо"].values.tolist().count('2 Ролла в подарок по Акции') * 2
    gifts_count = gifts_count + df_990[df_990['Номер чека'] == i]["Блюдо"].values.tolist().count('3 Ролла в подарок по Акции') * 3
    list.append({'check': i, 'sum' : sum_check, 'gift_count':gifts_count })

df = pd.DataFrame(list)
df['error'] = df['sum']//990
list_of_error = df[df['error']!=df['gift_count']].check.to_list()
df_990[df_990['Номер чека'].isin(list_of_error )].to_excel('неправильно выбитые подарочные роллы.xlsx')
```
